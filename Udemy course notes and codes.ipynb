```python
randn()
mylist=(10,100,1000,10000)

#watch out, no local vairables...scoping
for i in mylist:
    counter=0
    for jj in randn(i):
        if (jj < 1 and jj >-1):
                counter=counter+1
    answer=counter/i
    print('out', answer)
    
```
